```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
snb.set_theme()
```

```python
cars_sales_year = pd.read_csv("stacked_area_chart_data.csv")
cars_sales_year
```

```python
colors=["#011638","#7E2987","#ef2026"]
labels=["Diesel","Petrol","Gas"]
plt.figure(figsize=(13,7))
plt.stackplot(cars_sales_year["Year"],
              cars_sales_year["Diesel"],
              cars_sales_year["Petrol"],
              cars_sales_year["Gas"],
              colors=colors,
              edgecolor="none"
              )
plt.xticks(cars_sales_year["Year"], rotation=45)
plt.legend(labels=labels, loc="upper left")
plt.xlabel("Fuel engine types")
plt.ylabel("Number of cars")
plt.title("Popularity for engine fuel types (1982 - 2016)", fontweight="bold")
plt.show()
```
