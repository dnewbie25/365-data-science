```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="white")
from matplotlib.ticker import PercentFormatter
```

### Pareto chart
These bar and line chart is known also as Pareto chart

```python
data = pd.read_csv("bar_line_chart_data.csv")
data
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.bar(data["Year"],data["Participants"], color='k')
ax.set_ylabel("Number of participants", weight="bold")
ax.tick_params(axis="y",width=2,labelsize="large")
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(data["Year"], data["Python Users"], color="#b60000",
        marker="D")
ax1.set_ylabel("Python Users",color="#b60000", weight="bold")
ax1.tick_params(axis="y",colors="#b60000",width=2,labelsize="large")
ax.set_title("Python users (2012-2019)", fontsize=14)
plt.show()
```
