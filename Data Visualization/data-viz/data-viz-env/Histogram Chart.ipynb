```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
snb.set_theme()
```

```python
data = pd.read_csv("histogram_data.csv")
data
```

```python
plt.figure(figsize=(8,6))
plt.hist(data["Price"], bins=8, density=True,color="#108A99")
plt.title("Dist. of real state prices")
plt.xlabel("Price in thousands (000' $)")
plt.ylabel("Number of properties")
snb.set_theme(style="white")
plt.show()
```
