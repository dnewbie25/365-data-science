```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
snb.set_theme()
```

```python
data = pd.read_csv("histogram_data.csv")
data
```

```python
plt.figure(figsize=(8,6))
counts, bins, patches = plt.hist(data["Price"], bins=8, density=True,color="#108A99")
plt.title("Dist. of real state prices")
plt.xlabel("Price in thousands (000' $)")
plt.ylabel("Number of properties")
snb.set_theme(style="white")
bin_ranges = ["(%.1f, %.1f]" % (bins[i], bins[i+1]) for i in range(len(bins)-1)]
plt.xticks(bins[:-1],bin_ranges, rotation=45)
plt.show()
```
