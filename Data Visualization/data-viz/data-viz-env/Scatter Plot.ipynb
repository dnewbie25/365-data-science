```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
snb.set_theme()
```

```python
data = pd.read_csv("scatter_data.csv")
data
```

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(data["Area (ft.)"],
            data["Price"],
            alpha=0.6,
            c=data["Building Type"],
            cmap="viridis")
plt.xlabel("Area ft squared", fontsize=15)
plt.ylabel("Price in thousands", fontsize=15)
plt.legend(*scatter.legend_elements(),
           loc="upper left",
           title="Building type")
plt.title("Relationship Area vs Price in California")
plt.show()
```
