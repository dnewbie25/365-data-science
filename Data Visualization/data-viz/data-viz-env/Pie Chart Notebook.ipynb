```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
colors = snb.color_palette('colorblind')
```

```python
engine_fuel_type = pd.read_csv("pie_chart_data.csv")
engine_fuel_type
```

```python
plt.pie(engine_fuel_type["Number of Cars"],
        labels=engine_fuel_type["Engine Fuel Type"],
        autopct="%.2f%%",
        textprops={'size':'x-large', 'fontweight':'bold','rotation':30},
        colors=colors)
plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize=10)
plt.show()
```
