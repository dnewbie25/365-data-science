```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
data = pd.read_csv("scatter_plot_ii.csv")
data
```

```python
#plt.figure(figsize=(10,6))
sns.set_theme(rc={'figure.figsize':(9,6)})
sns.regplot(data=data,x="Budget",y="Sales",
            scatter_kws={'color':'blue'},
            line_kws={'color':'red'})
plt.xlabel("Ad expenditure in 000's $")
plt.ylabel("Sales in 000's $")
plt.show()
```

```python
sns.lmplot(data=data,x="Budget",y="Sales",height=10,
            scatter_kws={'color':'blue'},
            line_kws={'color':'red'})   
plt.show()
```
