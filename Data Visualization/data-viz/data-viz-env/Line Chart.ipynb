```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
snb.set_theme()
```

```python
spx_500 = pd.read_csv("line_chart_data.csv")
spx_500
```

```python
spx_500.info()
```

```python
spx_500["formatted date"] = pd.to_datetime(spx_500["Date"])
spx_500
```

```python
plt.figure(figsize=(20,8))
labels=["S&P 500","FTSE 100"]
plt.title("S&P Returns 2000 - 2010")
plt.plot(spx_500["formatted date"],spx_500["GSPC500"])
plt.plot(spx_500["formatted date"],spx_500["FTSE100"])
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(labels=labels, fontsize="large")
plt.show()
```

```python
spx_500_2008 = spx_500[(spx_500["formatted date"]>='2008-07-01') &
                      (spx_500["formatted date"]<='2008-12-31')]
spx_500_2008
```

```python
plt.figure(figsize=(20,8))
labels=["S&P 500","FTSE 100"]
plt.title("S&P Returns 2008")
plt.plot(spx_500_2008["formatted date"],spx_500_2008["GSPC500"])
plt.plot(spx_500_2008["formatted date"],spx_500_2008["FTSE100"])
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(labels=labels, fontsize="large")
plt.show()
```
