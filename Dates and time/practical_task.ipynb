```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import pytz
```

```python
sales = pd.read_csv("Sales-products-tz-mod.csv")
sales.head()
```

```python
modified_sales = sales.copy()
modified_sales["date"] = modified_sales["DateOfSale"]+" "+modified_sales["TimeOfSale"]
modified_sales["date"] = pd.to_datetime(modified_sales["date"],format="%d/%m/%Y %H:%M:%S")
modified_sales.head()
```

Here I find the unique values and find SGT as a timezone. Tha corresponds to Singapore. Since it was not in the pytz correct timezones, I modified that value to Singapore

```python
pytz.all_timezones
```

```python
modified_sales["TimeZone"].unique()
```

```python
modified_sales.loc[modified_sales["TimeZone"]=="SGT","TimeZone"] = "Singapore"
```

```python
modified_sales.head(19)
```

Create a function to put the timezone

```python
def assign_timezone(row):
  date = pd.to_datetime(row["date"])
  print(row["SaleID"])
  return date.tz_localize(row["TimeZone"])
```

```python
modified_sales["date"] = modified_sales.apply(assign_timezone,axis=1)
```

```python
modified_sales.head(20)
```

```python
sorted_df = modified_sales.copy()
sorted_df["date"] = pd.to_datetime(sorted_df["date"],format="%Y/%m/%d %H:%M:%S%z",utc=True)
```
