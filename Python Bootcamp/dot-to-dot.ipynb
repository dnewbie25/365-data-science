### Question 1

```python
# Enter your code here.
arr_X = []
arr_Y = []
with open('dot-to-dot.txt','r') as text:
  content = text.readlines()
  for line in content:
    separated = line.split(' ')
    arr_X.append(int(separated[0]))
    arr_Y.append(int(separated[1].strip()))

print(arr_X)
print(arr_Y)

import matplotlib.pyplot as plt

plt.plot(arr_Y, arr_X)
```
