# Random Walk

Using matplotlib to create a random walk graph, where each 1 moves the graph up, each -1 moves it down, and it is always moving to the right after each coin toss

```python
from random import choice, randint
import matplotlib.pyplot as plt

steps = [-1,1]
step_choice = choice(steps)
step_choice

walk_path = []
walk_path.append(randint(0,60)) # puts the starting point in a random place along the Y axis between 0 and 60
for step in range(0,1000):
  '''Adds a new random choice, -1 or 1, and appends it to the walk_path'''
  next_step = walk_path[-1]+choice(steps)
  walk_path.append(next_step)

plt.plot(walk_path)
plt.show()
```
